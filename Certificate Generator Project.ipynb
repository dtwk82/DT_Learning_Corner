```python
def certificate_generator(Name, Course, CourseLevel, Formatted_Date):

  pdf_file_name = "/content/drive/MyDrive/CRA - Project 01/Lesson 20 - Python Certificate Generator Project/certificates" + Name + "-" + Course + "-" + CourseLevel + ".pdf"

  c =  canvas.Canvas(pdf_file_name, pagesize = landscape(A4))

  template = "/content/drive/MyDrive/CRA - Project 01/Lesson 20 - Python Certificate Generator Project/certificate_template.jpg"

  c.drawImage(template, 0, 0, A4[1], A4[0])

  center_x = c._pagesize[0] / 2
  center_y = c._pagesize[1] / 2

  #1. Name
  c.setFont("Lora-Bold", 30)
  c.drawCentredString(center_x, center_y - 46, Name)

  #2. Course, CourseLevel
  c.setFont("Lora-Bold", 28)
  c.drawCentredString(center_x, center_y - 105, Course + " - " + CourseLevel)

  #3. Formatted_Date
  c.setFont("Lora-Bold", 17)
  c.drawCentredString(center_x + 190, center_y - 160, Formatted_Date)

  #4. Cert ID
  cert_id = "Cert ID: " + str(pd.Timestamp.now().timestamp()).replace(".","") # generating unique I based on current timestamp and later remove the dots to look better

  c.setFont("Lora-Bold", 12)
  c.drawCentredString(center_x + 266, center_y - 230, cert_id.upper())

  c.save()


certificate_generator("Sarah", "Python", "Beginner","02/07/2026")
```
